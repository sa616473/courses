```
import tensorflow as tf
```

```
import numpy as np
```

```
#Load and prep data
mnist = tf.keras.datasets.mnist

(X_train, y_train), (X_test, y_test) = mnist.load_data()

#convert the samples from integers to floating-point numbers
X_train, X_test = X_train / 255.0, X_test / 255.0
```

```
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten, Dropout,Softmax
```

```
#Building the neural network

model = Sequential()

model.add(Flatten(input_shape =(28,28)))
model.add(Dense(128, activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(10))
```

#log-odds
The model returns a vector of log odds

log-odds -> ln(odds) -> ln(P/(1-P))

odds = P/(1-P) Ex: .9/.1 = 9

Ex: ln(9) = 2.2

log-odds are inverse of sigmoid function



```
predictions = model(X_train[:1]).numpy()
predictions
```

```
#Converting the lod-odds to probabilities
tf.nn.softmax(predictions).numpy
```

```
#Caluclating the losses
loss_fn = tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True)
```

#loss = -log(Trueclass)
Ex: -log(1/10) ~= 2.3


```
loss_fn(y_train[:1], predictions).numpy()
#This should be approx 2.3
```

```
model.compile(optimizer='adam',
              loss=loss_fn,
              metrics=['accuracy'])
```

```
#Training data
model.fit(X_train, y_train, epochs=5)
```

```
#Predicting the data
model.evaluate(X_test,y_test, verbose=2)
```

```
#Proability model
proability_model = Sequential([model, Softmax()])
```

```
#Raw predictions
proability_model(X_test[:5])
```
